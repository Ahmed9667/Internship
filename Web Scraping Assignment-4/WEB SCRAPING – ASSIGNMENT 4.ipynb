# Q1
Scrape the details of most viewed videos on YouTube from Wikipedia. Url =
https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
ranks = []
names = []
artists = []
dates = []
views = []

#ranks
for i in driver.find_elements(By.XPATH , '//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody/tr/td[1]'):
    ranks.append(i.text)
    
#names
for i in driver.find_elements(By.XPATH , '//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody/tr/td[2]'):
    names.append(i.text)
    
#artists
for i in driver.find_elements(By.XPATH , '//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody/tr/td[3]'):
    artists.append(i.text)
    
#views
for i in driver.find_elements(By.XPATH , '//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody/tr/td[4]'):
    views.append(i.text)

#dates
for i in driver.find_elements(By.XPATH , '//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody/tr/td[5]'):
    dates.append(i.text)
    
videos = pd.DataFrame()
videos['Rank'] = ranks
videos['Name'] = names
videos['Artists'] = artists
videos['Views'] = views
videos['Date'] = dates
videos.set_index('Rank')
```

# Q2
Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1 ODI)
B) Series
C) Place
D) Date
E) Time

```python
driver = webdriver.Chrome()
driver.get('https://www.bcci.tv/')
```

```python
#get the page of international fixture
driver.find_element(By.XPATH , '/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a').click()
```

```python
title = []
teams = []
series = []
place = []
date = []
time = []

#teams
team1 = driver.find_elements(By.XPATH , '//div[@class="fixture-tab-inner row"]//div[@class="col-lg-4 col-md-6 col-sm-12 ng-scope"]//div[@class="match-tm-info teamLeft"]/div[2]')
team2 = driver.find_elements(By.XPATH , '//div[@class="fixture-tab-inner row"]//div[@class="col-lg-4 col-md-6 col-sm-12 ng-scope"]//div[@class="match-tm-info teamRight"]/div[2]')
for i in range(len(team1)):
    teams.append((team1[i].text,' vs ',team2[i].text))


#title
for i in driver.find_elements(By.XPATH , '//div[@class="fixture-tab-inner row"]//div[@class="col-lg-4 col-md-6 col-sm-12 ng-scope"]//div[@class="match-card-top"]/h5'):
    title.append(i.text)
    
#date
for i in driver.find_elements(By.XPATH , '//div[@class="fixture-tab-inner row"]//div[@class="col-lg-4 col-md-6 col-sm-12 ng-scope"]//div[@class="match-info"]/div[1]'):
    date.append(i.text)
    
#time
for i in driver.find_elements(By.XPATH , '//div[@class="fixture-tab-inner row"]//div[@class="col-lg-4 col-md-6 col-sm-12 ng-scope"]//div[@class="match-info"]/div[2]'):
    time.append(i.text)
    
#series
for i in driver.find_elements(By.XPATH , '//div[@class="fixture-tab-inner row"]//div[@class="col-lg-4 col-md-6 col-sm-12 ng-scope"]//div[@class="match-place ng-scope"]/span[1]'):
    series.append(i.text.strip(' -'))

#place
for i in driver.find_elements(By.XPATH , '//div[@class="fixture-tab-inner row"]//div[@class="col-lg-4 col-md-6 col-sm-12 ng-scope"]//div[@class="match-place ng-scope"]/span[2]'):
    place.append(i.text.strip(','))

fixture = pd.DataFrame()
fixture['Tilte'] = title
fixture['Teams'] = teams
fixture['Series'] = series
fixture['Place'] = place
fixture['Date'] = date
fixture['Time'] = time
fixture
```

# Q7
Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/ You
have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
url = 'https://www.imdb.com/list/ls095964455/'
page = requests.get(url)
source = page.content
soup = BeautifulSoup(source , 'lxml')
show = soup.find('div' , {'class' : 'lister-list'}).find_all('div' , {'class' : 'lister-item mode-detail'})
len(show)
```

```python
names = []
years = []
genres = []
runs = []
rates = []
votes = []
for i in range(len(show)):
    #find name
    name = show[i].find('h3' , {'class':'lister-item-header'}).find('a')
    names.append(name.text)
    
    #find year
    year = show[i].find('h3' , {'class':'lister-item-header'}).find('span',{'class':'lister-item-year text-muted unbold'})
    years.append(year.text.strip('()'))
    
    #run time
    time = show[i].find('p',{'class': 'text-muted text-small'}).find('span' , {'class' : 'runtime'})
    runs.append(time.text)
    
    #find genre
    genre = show[i].find('p',{'class': 'text-muted text-small'}).find('span' , {'class' : 'genre'})
    genres.append(genre.text.replace('\n' ,'').split('            ')[0])
    
    #find rates
    rate = show[i].find('div', {'class':'ipl-rating-widget'}).find('span',{'class':'ipl-rating-star__rating'})
    rates.append(rate.text)
    
    #find votes
    vote = show[i].find_all('p',{'class': 'text-muted text-small'})[2].find('span',{'name':'nv'})
    votes.append(vote.get('data-value'))
    
shows = pd.DataFrame(index = np.arange(1 , len(names)+1))
shows['Name'] = names
shows['Year Span'] = years
shows['Genre'] = genres
shows['Run Time'] = runs
shows['Rate'] = rates
shows['Votes'] = votes
shows
```

```python
driver = webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
#view datasets
driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]').click()
```

```python
url = []
for page in range(0,25):
    
    for i in driver.find_elements(By.XPATH , '//div[@class="h-12 w-12 rounded"]/a'):
        
        url.append(i.get_attribute('href'))
    
    driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]').click()
    
```

```python
len(url)
```

```python
name = []
characteristics = []
tasks = []
attri_type = []
instances = []
attribute = []
year = []

for i in url :
    driver.get(i)
    
    
    try:
        name.append(driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1').text)
        
        characteristics.append(driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p').text)
        
        tasks.append(driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p').text)
        
        attri_type.append(driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p').text)
        
        instances.append(driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p').text)
        
        attribute.append(driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p').text)
        
        year.append(driver.find_element(By.XPATH , '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2').text)
        
    except NoSuchElementException:
        name.append('-')
        characteristics.append('-')
        attri_type.append('-')
        tasks.append('-')
        instances.append('-')
        attribute.append('-')
        year.append('-')
```

```python
dataset = pd.DataFrame()
dataset['Name'] = name[:625]
dataset['Data Type'] = characteristics[:625]
dataset['Task'] = tasks[:625]
dataset['Attribute Type'] = attri_type[:625]
dataset['No of Instances'] = instances[:625]
dataset['No of Attribute'] = attribute[:625]
dataset['Year'] = year[:625]
dataset
```
