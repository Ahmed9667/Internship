```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
# Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
#have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
#jobs data

#first geth the page
driver = webdriver.Chrome()
driver.get('https://www.shine.com/')
```

```python
driver.find_element(By.CLASS_NAME , 'input').click()
```

```python
#Enter “Data Analyst” in “Job title, Skills” field 
driver.find_element(By.ID , 'id_q').send_keys('Data Analyst')
```

```python
#enter “Bangalore” in “enter the location” field
driver.find_element(By.ID , 'id_exp').send_keys('Bangalore')
```

```python
#then click search bottun
driver.find_element(By.CLASS_NAME , 'searchForm_btnWrap_advance__VYBHN').click()
```

```python
#Then scrape the data for the first 10 jobs results you get.
jobs=[]
title = driver.find_elements(By.XPATH , '//h2[@itemprop="name"]')
for i in title[0:10]:
    job = i.text
    jobs.append(job)

companies = []
company = driver.find_elements(By.XPATH , '//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company[0:10]:
    companies.append(i.text)

location = []
loc = driver.find_elements(By.XPATH , '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in loc[0:10]:
    location.append(i.text)

exp = []
exp_years = driver.find_elements(By.XPATH , '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in exp_years[0:10]:
    exp.append(i.text)

#no create dataframe
df=pd.DataFrame()
df['job title']= jobs
df['company']= companies
df['location']= location
df['experience'] = exp
df
```

```python
#Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
#6. Brand
#7. Product Description
#8. Price
```

```python
# go to flipkart webpage
driver = webdriver.Chrome()
driver.get('https://www.flipkart.com/')
```

```python
#Enter “sunglasses” in the search field where “search for products, brands and more” is written and click the search icon

search = driver.find_element(By.CLASS_NAME , '_3704LK')
search.send_keys('sunglasses')
```

```python
#press on search
driver.find_element(By.CLASS_NAME , 'L0Z3Pu').click()
```

```python
brands = []
Product_Description = []
sales = []

for page in (0,4):
    brand = driver.find_elements(By.XPATH , '//div[@class="_2WkVRV"]')
    for i in brand[0:100]:
        brands.append(i.text)
        
    production = driver.find_elements(By.XPATH , '//div[@class="_2B099V"]/a[1]') #a[1] the first tag name a
    for i in production[0:100]:
        Product_Description.append(i.get_attribute('title')) 
        
    price = driver.find_elements(By.XPATH , '//div[@class="_25b18c"]/div[@class="_30jeq3"]')
    for i in price[0:100]:
        sales.append(i.text)
        
    
    next_bottun = driver.find_element(By.XPATH , '//a[@class="_1LKTO3"]')
    next_bottun.click()
    time.sleep(2)
```

```python
mobiles1 = pd.DataFrame()
mobiles1['brands'] = brands
mobiles1['products'] = Product_Description
mobiles1['prices'] = sales
mobiles1
```

```python
#Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
# Term of office, Remarks) from https://www.jagranjosh.com/.
```

```python
#get the page

driver = webdriver.Chrome()
driver.get('https://www.jagranjosh.com/')
```

```python
#click on GK

driver.find_element(By.XPATH , '/html/body/div/header/nav/div/div/div[3]/ul/li[3]/a').click()
```

```python
# press on list of prime ministers of india
driver.find_element(By.XPATH , '/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a').click()
```

```python
import numpy as np
names = []
lives = []
offices = []
remarks = []


name = driver.find_elements(By.XPATH , '//tbody/tr/td[2]')
life  = driver.find_elements(By.XPATH , '//tbody/tr/td[3]')
in_office = driver.find_elements(By.XPATH , '//tbody/tr/td[4]')
remark = driver.find_elements(By.XPATH , '//tbody/tr/td[5]')

#find names of prime minister :
for i in range(0,19) :  #here we used range(0,19) because '//tbody/tr/td[2]' gives us 20 values
    names.append(name[i].text)
    
#find boorn and dead :
for i in life:
    lives.append(i.text)
    
#find in offices for every minister:
for i in in_office:
    offices.append(i.text)
    
#find remarks:
for i in remark:
    remarks.append(i.text)
    
ministers = pd.DataFrame(index= np.arange(1,20))
ministers['PM NAME'] = names
ministers['Born-Dead'] = lives
ministers['Term of office'] = offices
ministers['Remark'] = remarks
ministers  
```
